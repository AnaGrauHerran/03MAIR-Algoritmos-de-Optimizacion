```python
# Dado un equipo calcula su indice correspondiente para matriz de audiencias
def calculo_indice(equipo): # O(n), recorrer listado de todos los equipos hasta encontrar equipo)
    if equipo in equipos_A:
        return 0
    if equipo in equipos_C:
        return 2
    return 1

# Dado un equipo, determina si esta clasificado como equipo A, B o C
def calculo_clasificacion(equipo): # O(n), recorrer listado de todos los equipos hasta encontrar equipo)
    if equipo in equipos_A:
        return 'A'
    if equipo in equipos_C:
        return 'C'
    return 'B'

# Dado un equipo, calcula su audiencia ideal (S-20)
def audiencia_partido(partido, matriz_audiencias): # O(n) 2 calculos_indice() + busqueda en matriz de audencia que tiene un tamaño constante de 9.
    indice_1 = calculo_indice(partido[0])
    indice_2 = calculo_indice(partido[1])
    return matriz_audiencias[indice_1][indice_2]

# Dado un listado de partidos de la jornada, devuelve una lista con sus audiencias ideales (S-20) en orden
def calculo_audiencias_ideales(jornada, matriz_audiencias): # O(n^2) for loop de tamaño n con operación O(n) dentro => n * n = O(n^2)
    audiencias = []
    for partido in jornada:
        audiencias.append(audiencia_partido(partido, matriz_audiencias))
    return audiencias
```

```python
import pandas as pd

# Dado un listado de partidos (audiencias) en orden, se calcula la audiencia total de la jornada
def calculo_audiencia_real(partidos, pesos): # O(n) 1 for loop tamaño n
    audiencia = sum([partidos[i] * pesos[i] for i in range(len(partidos))])
    return round(audiencia, 2)

# Metodo recursivo para calcular un listado con todas las posibles combinaciones de jornadas
def generar_soluciones(partidos, soluciones): # O(n+1!) => Explicacion luego en el apartado correspondiente
    # si las soluciones no son vacias y el tamaño de solucion es el de la jornada salimos
    if soluciones and len(soluciones[0])==len(partidos):
        return soluciones
    nuevas_soluciones = []
    for solucion in soluciones:
        partidos_restantes = list(set(partidos) - set(solucion))
        for partido in partidos_restantes:
            nueva_solucion = solucion + [partido]
            nuevas_soluciones.append(nueva_solucion)
    return generar_soluciones(partidos, nuevas_soluciones)

# Dado un listado con el orden que deben de tener los partidos, se ordenan las audiencias
#(se usa para luego calcular la audiencia de dicha posible solucion)
def obtener_audiencia_solucion(indices, partidos): # O(n^2), for loop * localizacion del elemento indice en partidos
    solucion = []
    for indice in indices:
        solucion.append(partidos[indice])
    return solucion

# Dado un listado con el orden de los partidos de la solucion, se construye el "partido" y su clasificacion para imprimir
def obtener_equipos(indices, jornada): # O(n^2), for loop * localizacion del elemento indice en jornada
    solucion = []
    clasificacion = []
    for indice in indices:
        partido = jornada[indice]
        solucion.append(partido[0] + '-' + partido[1])
        clasificacion.append(calculo_clasificacion(partido[0]) + '-' + calculo_clasificacion(partido[1]))
    return [solucion, clasificacion]

# Imprime la solucion de la jornada
def imprimir_solucion(solucion, clasificacion, horarios): # O(n) imprimir listados de tamaño n
    data = { 'horarios': horarios, 'partidos': solucion, 'clasificacion': clasificacion}
    print(pd.DataFrame.from_dict(data))
    
def organizar_partido_fuerza_bruta(jornada, pesos, horarios, matriz_audiencias):
    # Paso 1: iniciar variables
    size = len(jornada)
    mejor_solucion = []
    mejor_audiencia = 0

    # Paso 2: calculo audiencias base de partidos de la jornada
    partidos = calculo_audiencias_ideales(jornada, matriz_audiencias)

    # Paso 3: construir todas las soluciones con los indices
    inicializacion_soluciones = []
    indices = list(range(len(partidos)))
    for indice in indices:
        inicializacion_soluciones.append([indice])

    todas_soluciones = generar_soluciones(indices, inicializacion_soluciones)

    # Paso 4: encontrar mejor solucion entre todas las soluciones
    for solucion_indices in todas_soluciones:
        solucion = obtener_audiencia_solucion(solucion_indices, partidos)
        # calculo de audiencia
        audiencia_actual = calculo_audiencia_real(solucion, pesos)
        # calcular si es optima
        if audiencia_actual > mejor_audiencia:
            mejor_solucion = solucion_indices
            mejor_audiencia = audiencia_actual
    
    # Paso 5: devolver resultados  
    print("Mejor solucion: ")
    jornada_solucion, clasificacion_solucion = obtener_equipos(mejor_solucion, jornada)
    imprimir_solucion(jornada_solucion, clasificacion_solucion, horarios)
    print("----------------------------------------------")
    print("Mejor audiencia: ")
    print(str(round(mejor_audiencia,2)) + " millones de espectadores")
```

Calcula la complejidad del algoritmo por fuerza bruta

**Respuesta**

El algoritmo por fuerza bruta tiene 3 parámetros que corresponden a tres listas y una matriz de tamaño constante 9. Todas las listas (en nuestro caso) tienen el mismo tamaño n = número de partidos en la jornada ya que para nuestro algoritmo de fuerza bruta asumimos que los horarios disponibles y los partidos a disputar en la jornada coinciden.

Paso 1: Cálculo de size = n operaciones

        Asignaciones = 2 operaciones
       
Paso 2: Calculo_audiencias_ideales = O(n^2)

Paso 3: Asignación = 1 operación

        Inicializar índices = O(n) (generar lista de índices + iniciar listado soluciones)
        
        Generar_soluciones = O(n+1!)
       
Paso 4: Loop sobre todas las operaciones posibles = O(n!) * (obtener_audiencia_solucion + calculo_audiencia_real + 3)

        = O(n!) * (O(n^2)+O(n)+3)
        
        = O((n+2)!)

Paso 5: Imprimir mensaje = 1 operación

        Obtener equipos = O(n^2)
        
        Imprimir solucion = O(n)
        
        Imprimir audiencia = 3 operaciones
       
La complejidad del algoritmo de fuerza bruta es O((n+2)!)

-----
Veamos con mas detalle la complejidad del método generar_soluciones

En la primera iteración se hacen O(n^2) operaciones ( en realidad O(n * n-1))

En la siguiente iteracion, la lista de soluciones es del orden O(n^2) por lo que hago O(n^3) operaciones (en realidad O(n * n-1 * n-2))

....

En la última iteración (antes de devolver las soluciones), he hecho n bucles (o iteraciones), y el bucle con más operaciones es el último con un orden de O(n!)

Por ello se puede concluir que su orden de complejidad es de O(n+1)!


(*) Diseña un algoritmo que mejore la complejidad de algoritmo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

**Respuesta**

Mejoraremos el algoritmo de fuerza bruta usando un algoritmo voraz para encontrar la solución. De esta manera será suficiente hacer recorridos sobre la lista de la jornada y los pesos para encontrar la mejor solución u opción en cada "iteración". Este método no requiere generar todas las soluciones posibles y evaluarlas, bastaría con crear un método de ordenacion tipo quick_sort (complejidad O(n log n) o en el peor de los casos O(n^2)) que ordene los partidos en función de su posible audiciencia. Usar el mismo método para ordenar los horarios en función de sus pesos. Luego bastaría con "emparejar" los mejores horarios con los mejores partidos.
Este método mejoraría significativamente el diseño de fuerza bruta bajando la complejidad de O(n+2!) a una complejidad no superior a O(n^3).

```python
# Dadas las audiencias base de los partidos de la jornada
# crea una lista con los indices de los partidos
# donde los primeros son los de mayor audiencia y los ultimos los de menor audiencia.
def ordenar_partidos_quick_sort(partidos, indices, pivote=None):
    # Calcular tamaño de la lista
    size = len(indices)
    
    # Si la lista tiene un elemento
    # o es vacía devolverla tal cual
    if size == 1 or size == 0:
        return indices
    # Si la lista tiene dos elementos se puede
    # ordenar facilmente con la función máximo y mínimo
    if size == 2:
        i = indices[0]
        j = indices[1]
        A = partidos[i]
        B = partidos[j]
        if A > B:
            return [i, j]
        else:
            return [j, i]
    
    # Crear listas auxiliares
    izquierda = []
    derecha=[]
    pivote_izquierdo=0
    pivote_derecho=0
    cantidad_izquierda=0
    cantidad_derecha=0

    # Calcular pivote
    if not pivote:
        pivote = (partidos[indices[0]]+ partidos[indices[1]]+ partidos[indices[2]])/3

    # Para cada elemento de la lista, si es mayor al pivote
    # va a la lista de la izquierda y la derecha si es menor
    for indice in indices:
        partido = partidos[indice]
        if partido > pivote:
            izquierda.append(indice)
            pivote_izquierdo += partido
            cantidad_izquierda +=1
        else:
            derecha.append(indice)
            pivote_derecho += partido
            cantidad_derecha += 1
    
    # En el caso de que haya tres o más números repetidos en la lista
    # una de las listas auxiliares es 0 por lo que se devuelve la lista
    if cantidad_izquierda==0:
        return derecha
    elif cantidad_derecha ==0:
        return izquierda
    
    # Se hace recursión sobre las dos listas auxiliares
    # la solución es la unión de esas dos listas
    return ordenar_partidos_quick_sort(partidos, izquierda, pivote_izquierdo/cantidad_izquierda) + ordenar_partidos_quick_sort(partidos, derecha, pivote_derecho/cantidad_derecha)

# Metodo voraz para asignar los horarios(contando las restricciones correspondientes)
# Este metodo permite asignar más de un partido a un horario
def asignacion_horarios(indices_ordenados, partidos, pesos):
    # iniciar variables
    size_pesos = len(pesos)
    size_indices = len(indices_ordenados)
    mejor_solucion = [[] for _ in range(size_pesos)]
    mejor_audiencia = 0
    # indices_ordenados son los indices de los partidos ordenados de mayor a menor audiencia
    for i in range(size_indices):
        #obtenemos el partido a asignar
        indice = indices_ordenados[i]
        partido = partidos[indice]
        
        # inicializamos las variables para calcular su horario solucion
        audiencia = 0
        solucion = 0
        
        # para los dos ultimos 'partidos' si no hay partido en lunes o viernes se le asigna el lunes o viernes (requisito)
        es_partido_ultimo = (i==(size_indices - 1) or i == (size_indices -2))
        es_viernes_vacio = not mejor_solucion[0]
        es_lunes_vacio = not mejor_solucion[size_pesos - 1]
        es_requisito_vacio = es_viernes_vacio or es_lunes_vacio
            
        # si estamos en condiciones de requisito
        if es_partido_ultimo and es_requisito_vacio:
            if es_viernes_vacio:
                solucion = 0
                audiencia = partido * pesos[0]
            else:
                solucion = size_pesos - 1
                audiencia = partido * pesos[solucion]
        # el resto de los casos
        else:
            # elegir mejor opcion
            for j in range(size_pesos):
                peso = pesos[j]
                if not mejor_solucion[j]:
                    posible_audiencia = partido * peso
                else:
                    # compartir horario no añadiría valor a la audiencia total ya que, al estar las audiencias ordenadas de
                    # mayor a menor, los partidos comparten audiencia total en la franja del partido con mayor audiencia
                    # si se cambia el programa para otro sistema de combinación de horarios, esto se sustituiría por la
                    # audiencia añadida por el partido a la audiencia total
                    posible_audiencia = 0
                if posible_audiencia > audiencia:
                    audiencia = posible_audiencia
                    solucion = j
        mejor_solucion[solucion].append(indice)
        mejor_audiencia += audiencia
    return [mejor_solucion, mejor_audiencia]

# Dado un listado con el orden de los partidos de la solucion, se construye el "partido", su clasificacion
# para imprimir y su horario. Este método ha sido adaptado para tolerar varios partidos en un mismo horario
def obtener_equipos_coincidentes(indices, jornada, horarios): # O(n^3), 2 for loop * localizacion del elemento indice en jornada
    solucion = []
    clasificacion = []
    horarios_jornada = []
    for i in range(len(indices)):
        horario = horarios[i]
        for indice in indices[i]:
            partido = jornada[indice]
            solucion.append(partido[0] + '-' + partido[1])
            clasificacion.append(calculo_clasificacion(partido[0]) + '-' + calculo_clasificacion(partido[1]))
            horarios_jornada.append(horario)
    return [solucion, clasificacion, horarios_jornada]

def organizar_partido_voraz(jornada, pesos, horarios, matriz_audiencias):
    # Paso 1: iniciar variables
    size = len(jornada)
    mejor_solucion = []
    mejor_audiencia = 0

    # Paso 2: calculo audiencias base de partidos de la jornada
    partidos = calculo_audiencias_ideales(jornada, matriz_audiencias)
    
    # Paso 3: ordenar partidos en funcion de sus audiencias.
    indices_ordenados = ordenar_partidos_quick_sort(partidos, list(range(size)))

    # Paso 4: asignacion de horarios
    mejor_solucion, mejor_audiencia = asignacion_horarios(indices_ordenados, partidos, pesos)

    # Paso 5: devolver resultados  
    print("Mejor solucion: ")
    jornada_solucion, clasificacion_solucion, horarios_solucion = obtener_equipos_coincidentes(mejor_solucion, jornada, horarios)
    imprimir_solucion(jornada_solucion, clasificacion_solucion, horarios_solucion)
    print("----------------------------------------------")
    print("Mejor audiencia: ")
    print(str(round(mejor_audiencia,2)) + " millones de espectadores")
```
