```python
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución(Isabel Vázquez)

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
   
    def wrapper(*args, **kwargs):       
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    
    return wrapper
```

```python
import random

N = 500
LISTA_2D = [ (random.randrange(1,N*10), random.randrange(1,N*10)) for _ in range(N)]
```

```python
import math

############
## MEJORA ##
############
# 1.- Verificar que A y B son los dos 1_D o del mismo tamaño
# Notar que si B es de menor dimensión que A puede generar error en la función distancia <- Arreglado

def is_1D_number(punto):
    tipo = type(punto)
    if tipo is int or tipo is float:
        return True
    return False

def is_same_dimension(punto_A, punto_B):
    return len(punto_A) == len(punto_B)

def distancia(punto_A, punto_B):
    # Mejora 1a.
    if is_1D_number(punto_A) and is_1D_number(punto_B):
        return abs(punto_B-punto_A)
    # Mejora 1b.
    elif is_same_dimension(punto_A, punto_B):
        return math.sqrt(sum([(punto_A[i]-punto_B[i])**2 for i in range(len(punto_A))]))
    else:
        print(punto_A)
        print(punto_B)
        print("Puntos A y B deben tener la misma dimension")
         
```

```python
# Problema dos puntos más cercanos
# Método Fuerza bruta
import sys

############
## MEJORA ##
############
# 2.- Utilizar la variable máxima del sistema como cota máxima[sys.maxsize]
# 3.- Longitud y distancias calculadas una sola vez en el programa
# 4.- Verificar que la lista tiene tamaño superior a 2 (sino, se devuelve la lista). Esta mejora asegura buen funcionamiento
# en el método distancia_divide_y_venceras cuando la "frontera" es vacía o tiene un sólo punto

def distancia_fuerza_bruta(lista):
    # inicializar las variables auxiliares
    # Mejora 2.
    mejor_distancia = sys.maxsize
    punto_A,punto_B=(),()
    
    # Mejora 3a.
    size = len(lista)
    # Mejora 4.
    if size==0 or size==1 or size==2:
        return lista
    
    # para todos los elementos de la lista
    for i in range(size):
        # se calcula la distancia con los elementos "restantes" de la lista
        for j in range(i+1, size):
            # Mejora 3b.
            distancia_actual = distancia(lista[i],lista[j])
            # si la distancia es menor que las previas
            if distancia_actual < mejor_distancia:
                # los puntos se convierten en candidatos a solución
                punto_A,punto_B=lista[i],lista[j]
                # la mejor (menor) distancia es la actual
                mejor_distancia = distancia_actual
    return[punto_A,punto_B]

@calcular_tiempo
def distancia_fuerza_bruta_decorada(lista):
    return distancia_fuerza_bruta(lista)

distancia_fuerza_bruta_decorada(LISTA_2D)
    
```

```python
############
## MEJORA ##
############
# 5- Implementar quick_sort en la primera variable
import sys

def quick_sort_on_x(lista, pivote=None):
    # Calcular tamaño de la lista
    size = len(lista)
    
    # Si la lista tiene un elemento
    # o es vacía devolverla tal cual
    if size == 1 or size == 0:
        return lista
    # Si la lista tiene dos elementos se puede
    # ordenar facilmente con la función máximo y mínimo
    if size == 2:
        A = lista[0][0]
        B = lista[1][0]
        if A < B:
            return [lista[0], lista[1]]
        else:
            return [lista[1], lista[0]]
    
    # Crear listas auxiliares
    izquierda = []
    derecha=[]
    pivote_izquierdo=0
    pivote_derecho=0
    cantidad_izquierda=0
    cantidad_derecha=0

    # Calcular pivote
    if not pivote:
        pivote = (lista[0][0]+ lista[1][0]+ lista[2][0])/3

    # Para cada elemento de la lista, si es menor al pivote
    # va a la lista de la izquiera y la derecha si es mayor
    for elemento in lista:
        if elemento[0] < pivote:
            izquierda.append(elemento)
            pivote_izquierdo += elemento[0]
            cantidad_izquierda +=1
        else:
            derecha.append(elemento)
            pivote_derecho += elemento[0]
            cantidad_derecha += 1
    
    # En el caso de que haya tres o más números repetidos en la lista
    # una de las listas auxiliares es 0 por lo que se devuelve la lista
    if cantidad_izquierda==0:
        return derecha
    elif cantidad_derecha ==0:
        return izquierda

    # Se hace recursión sobre las dos listas auxiliares
    # la solución es la unión de esas dos listas
    return quick_sort_on_x(izquierda, pivote_izquierdo/cantidad_izquierda) + quick_sort_on_x(derecha, pivote_derecho/cantidad_derecha)

print(quick_sort_on_x(LISTA_2D))
```

```python
############
## MEJORA ##
############
# 6.- Usar quick_sort en la primera variable en vez de sorted
# 7.- Buscar soluciones en la franja entre las dos listas divididas 

# Problema dos puntos más cercanos
# Método Divide y Vencerás

def distancia_divide_y_venceras(lista):
    # Si hay pocos elementos en la lista usamos fuerza Bruta
    size = len(lista)
    if size < 100:
        return distancia_fuerza_bruta(lista)
    
    # Ordenamos la lista y dividimos la lista por la mitad
    # Division entera
    size_mitad = size//2
    # Mejora 6.
    lista_ordenada = quick_sort_on_x(lista)
    lista_izquierda = lista_ordenada[:size_mitad] 
    lista_derecha = lista_ordenada[size_mitad:]
    
    # calculamos los puntos más cercanos en las listas laterales
    puntos_lista_izquierda = distancia_divide_y_venceras(lista_izquierda)
    puntos_lista_derecha = distancia_divide_y_venceras(lista_derecha)

    # El candidato a solución es el mejor de las dos listas
    candidatos_laterales = distancia_fuerza_bruta(puntos_lista_izquierda + puntos_lista_derecha)
    distancia_candidato = distancia(candidatos_laterales[0], candidatos_laterales[1]) 

    # Calculamos el área frontera (por si la solución es un punto de la "izquierda" con uno de la "derecha")
    # Para ello extraemos de la lista_izquierda y de la lista_derecha aquellos puntos cuya coordenada x estén a una distancia
    # menor o igual a la distancia_candidato de la frontera. Es decir buscamos en los puntos que estén en una franja alrededor de
    # la frontera con radio distancia_candidato
    # Mejora 7.
    frontera = []
    pivote = lista[size_mitad]
    for elemento in lista_derecha:
        if distancia(elemento[0], pivote[0])<distancia_candidato:
            frontera.append(elemento)
        else:
            # como la lista ya está ordenada salimos del bucle
            break
    for i in range(size_mitad - 1 , 0, -1):
        elemento = lista_izquierda[i]
        if distancia(elemento[0], pivote[0])<distancia_candidato:
            frontera.append(elemento)
        else:
            # como la lista ya está ordenada salimos del bucle
            break
            
    candidato_frontera = distancia_divide_y_venceras(frontera)
    return distancia_fuerza_bruta(candidato_frontera+candidatos_laterales)

@calcular_tiempo
def distancia_divide_y_venceras_decorada(lista):
    return distancia_divide_y_venceras(lista)

distancia_divide_y_venceras_decorada(LISTA_2D[:200])
```

```python
############
## MEJORA ##
############
# 8.- Escribir método que dado punto de salida y de destino devuelva
# de manera explicita el precio y la ruta óptima

# Problema Viaje por el rio

tarifas_viaje = [
[  0,   5,   4,   3, 999, 999, 999],
[999,   0, 999,   2,   3, 999,  11],
[999, 999,   0,   1, 999,   4,  10],
[999, 999, 999,   0,   5,   6,   9],
[999, 999, 999, 999,   0, 999,   4],
[999, 999, 999, 999, 999,   0,   3],
[999, 999, 999, 999, 999, 999,   0]
]

def precios(tarifas):
    size = len(tarifas[0])

    # matriz de precios optimizados
    precios = [ [999]*size for i in [999]*size ]
    
    # rutas optimizadas
    ruta = [ [""]*size for i in [999]*size]

    # para cada ruta
    for i in range(size-1):
        # cada ruta restante
        for j in range(i+1, size):
            # candidato (i,j)
            minimo = tarifas[i][j]
            ruta[i][j] = i
            # calculo alternativa con intermediario
            for k in range(i,j):
                intermedio = precios[i][k] + tarifas[k][j]
                if  intermedio < minimo:
                    minimo = intermedio
                    ruta[i][j] = k
            precios[i][j] = minimo
            
    return precios, ruta
            
precio, ruta = precios(tarifas_viaje)
print(precio)
print(ruta)

# Calcular la mejor ruta del punto A al punto B
# el resultado incluye los puntos A y B en su solución
# así como todas las paradas intermedias.
# Mejora 8
def calcular_mejor_ruta(salida, llegada):
    # calculo de rutas y precios optimos
    precio, ruta = precios(tarifas_viaje)
    
    # Extraer opción optima desde nuestra salida
    opciones_optimas = ruta[salida]
    
    # Calcular el camino optimo(extraer el camino optimo)
    camino = opciones_optimas[llegada]
    solucion = [llegada]
    while camino != salida:
        solucion = [camino] + solucion
        camino = opciones_optimas[camino]

    # Imprimir la solución
    print("La ruta más barata es:")
    print([salida] + solucion)
    print("El precio es: " + str(precio[salida][llegada]))

calcular_mejor_ruta(0,6)
calcular_mejor_ruta(1,5)
```
